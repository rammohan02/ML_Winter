# Q2

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#a) Generating vector
mean = [1, 2, 1]
cov = [[5, 0.8, -0.3], [0.8, 3, 0.6], [-0.3, 0.6, 4]]

mnd = np.random.multivariate_normal(mean, cov, size=300)
mnd
```

```python
#b) Transposing the vector to get x1, x2, x3
mndt = mnd.T
x1 = mndt[0]
x2 = mndt[1]
x3 = mndt[2]
```

```python
plt.scatter(x1, x2)
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
plt.scatter(x1, x3)
plt.xlabel("X1")
plt.ylabel("X3")
plt.show()
```

```python
plt.scatter(x2, x3)
plt.xlabel("X2")
plt.ylabel("X3")
plt.show()
```

Data around mean points is very dense when compared to points we go farther

```python
#c) Calculating mean and covariance from generated vectors
mean_ = mnd.mean(axis=0)
cov_ = np.cov(mndt)
print("Mean: ", end="")
print(mean_)
print("Covariance: ", end="")
print(cov_)
```

# Q3

```python
from PIL import Image
```

```python
# Reading the image and displaying
img = Image.open("Lena.png")
img.show()
```

```python
# Converting into array
img_arr = np.array(img)
plt.imshow(img_arr)
```

```python
#a) Extracting 32x32 patch and displaying
img_32x32 = img_arr[100:132, 100:132, 0]
plt.imshow(img_32x32)
print(img_32x32)
```

```python
#b) Max in each row and column
max_row = np.amax(img_32x32, axis=1)
max_col = np.amax(img_32x32, axis=0)
print("Max in each row: ", end="")
print(max_row)
print("Max in each column: ", end="")
print(max_col)
```

```python
#c) Finding cosine of the patch
img_cos = np.cos(img_32x32)
print(img_cos)
```

# Q4

```python
import pandas as pd
```

```python
# Reading data
data4 = pd.read_csv("https://www.statlearning.com/s/Auto.csv")
data4
```

```python
# Finding mean using groupby
man_mpg = data4.groupby('cylinders').mean()['mpg']
man_mpg
```

```python
# Scallter of mpg vs displacement
plt.scatter(data4['mpg'], data4['displacement'])
plt.xlabel("MPG")
plt.ylabel("Displacement")
plt.title("MPG vs Displacement")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

ax = plt.figure().add_subplot(projection="3d")

# Make data.
X = np.outer(np.linspace(-3, 3, 32), np.ones(32))
Y = X.copy().T # transpose
Z = (np.sin(X **2) + np.cos(Y **2) )

# Plot the surface.
surf = ax.plot_surface(X, Y, Z)

# Customize the z axis.
ax.set_zlim(-2, 2)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

plt.show()
```
